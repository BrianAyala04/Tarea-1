```python
#Ejercicio 3
def salario(hora,costo):
    costo=150
    sd=costo*hora
    print("El salario de un dia es: ",sd)
    ss=5*(costo)*hora
    print("El salario semanal es: ",ss)
    sm=20*(costo)*hora
    print("El salario mensual es: ",sm)
hora=float(input("Numero de horas trabajadas: "))
salario(hora,150)
```
